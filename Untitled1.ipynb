```python

#Lab 2
import pandas as pd
import numpy as np

from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)

print(f"\nüìä DATASET SHAPE:")
print(f"   Rows: {df.shape[0]}")
print(f"   Columns: {df.shape[1]}")
print("=" * 50)

print("\nüëÄ FIRST 5 ROWS OF DATA:")
display(df.head())
print("=" * 50)

print("\nüìã COLUMN NAMES:")
for i, col in enumerate(df.columns, 1):
    print(f"   {i}. {col}")
print("=" * 50)

print("\nüîß DATA TYPES:")
print(df.dtypes)
print("=" * 50)

print("\nüéØ TARGET VARIABLE 'made_shot':")
if 'made_shot' in df.columns:
    target_counts = df['made_shot'].value_counts()
    print(f"   Made shots (TRUE):  {target_counts.get(True, 0)}")
    print(f"   Missed shots (FALSE): {target_counts.get(False, 0)}")
    print(f"   Success rate: {target_counts.get(True, 0) / len(df) * 100:.1f}%")
else:
    print("   ‚ùå Column 'made_shot' not found. Available columns:")
    for col in df.columns:
        print(f"      - {col}")
print("=" * 50)

print("\n‚ö†Ô∏è  MISSING VALUES:")
missing = df.isnull().sum()
if missing.sum() == 0:
    print("   No missing values found! ‚úÖ")
else:
    for col, count in missing.items():
        if count > 0:
            print(f"   {col}: {count} missing values")
print("=" * 50)

print("\nüìà BASIC STATISTICS FOR NUMERICAL COLUMNS:")
print(df.describe())
print("=" * 50)

print("\nüéâ DATA INSPECTION COMPLETE!")
```
